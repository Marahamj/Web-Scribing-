```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import json
import csv
import os
import shutil
try:
  os.mkdir("car_images")
except FileExistsError:
  shutil.rmtree("/content/car_images")
  os.makedirs("car_images")
filecsv = open('OpenSouqnew.csv', 'w', encoding='utf8')
file = open('OpenSouqnew.json', 'w', encoding='utf8')
# Set the URL you want to webscrape from
url = 'https://jo.opensooq.com/ar/%D8%B3%D9%8A%D8%A7%D8%B1%D8%A7%D8%AA-%D9%88%D9%85%D8%B1%D9%83%D8%A8%D8%A7%D8%AA/%D8%B3%D9%8A%D8%A7%D8%B1%D8%A7%D8%AA-%D9%84%D9%84%D8%A8%D9%8A%D8%B9?page='

file.write('[\n')
data = {}
csv_columns = ['name', 'price', 'img', 'city', 'area']
writer = csv.DictWriter(filecsv, fieldnames=csv_columns)
writer.writeheader()

for page in range(1,10):
    print('---', page, '---')

    # r = requests.get(url + str(page))

    # headers = {
    # 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0 Safari/537.36'
    # }
    # r = requests.get(url + str(page), headers=headers)
    r = requests.get(url + str(page))


    print(url + str(page))
    soup = BeautifulSoup(r.content, "html.parser")

    ancher=soup.find('div',{'class':'sc-88230b31-1 fitXRA'}).find_all('a')
    for pt in ancher:
      if pt.find('h2', {'class': 'breakWord trimTwoLines font-20 me-auto'}):
          name = pt.find('h2', {'class': 'breakWord trimTwoLines font-20 me-auto'})
          print("name",name.get_text().strip())
          price = pt.find('div', {'class': 'priceColor bold alignSelfCenter font-18 ms-auto'})
          price=price.get_text()
          price=price[:-6]
          print("price",price)
          img = pt.find('div', {'class': 'postImgCont relative radius-8 overflowHidden me-16 grayBg'}).find('img')['src']
          print("img phat",img)
          # استخراج اسم الصورة مع تصحيح الامتداد
          image_name = name.get_text().replace('/', '').replace('\\', '').replace(' ', '_').strip()

          # تحديد الامتداد الصحيح (قبل الأخير)
          image_url_parts = img.split('/')
          last_part = image_url_parts[-1]  # مثل: 27bc211fb7...jpg.webp
          ext_parts = last_part.split('.')  # ['27bc211fb7...', 'jpg', 'webp']

          # نأخذ الامتداد الثاني من الأخير (.jpg)
          if len(ext_parts) >= 2:
            true_ext = ext_parts[-2]
          else:
            true_ext = 'jpg'  # fallback

          image_filename = f"{image_name}.{true_ext}"

          # مسار الحفظ
          image_path = os.path.join("car_images", image_filename)

          # حفظ الصورة
          urllib.request.urlretrieve(img, image_path)

          text_location = pt.find('div', {'class': 'flex alignItems gap-5 darkGrayColor'})
          city, area = [x.strip() for x in text_location.get_text().split(",")]
          print("city",city)
          print("area",area)
          writer.writerow({'name': name.text, 'price': price,
                            'img': img,'city':city,'area':area})
          data['name'] = name.text.strip()
          data['price'] = price
          data['img'] = img
          data['city'] = city
          data['area'] = area
          json_data = json.dumps(data, ensure_ascii=False)
          file.write(json_data)
          file.write(",\n")
          print("-----------------")
    #       time.sleep(3)
    # time.sleep(3)
file.write("\n]")
filecsv.close()
file.close()
print(len(os.listdir('/content/car_images')), "car images are extracted")
```
